### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
def y(x):
    result = x*x
    return result
print(y(4)) 

# Generate x and y values for the plot
x_values = np.linspace(-4, 4, 100)
y_values = [y(x) for x in x_values]
# Plot the numerical result against the true normal distribution
plt.figure()
plt.plot(x_values, y_values, label='Plot y = x*x')
#plt.plot(x_values, st.norm.pdf(x_values, loc=1, scale=2), label='True Normal Distribution')
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot y = f(x)')
plt.show()
```
