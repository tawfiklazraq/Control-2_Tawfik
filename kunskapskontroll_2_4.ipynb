```python
# Importing Pandas. 
import pandas as pd
```

```python
houses = pd.read_csv("housing.csv")
houses.head(120)
#print(houses)
print('Number of houses in total is:', len(houses))
```

```python
print('mean for median_income for houses is:', houses[['median_income']].mean())
print('')
print('mean for median_house_value for houses is:',houses[['median_house_value']].mean())
print('')
print('mean for housing_mean_age for houses is:', houses[['housing_median_age']].mean())
print('')
print('mean for total_rooms for houses is:', houses[['total_rooms']].mean())
```

```python
######################################################################################
## In my study I compare statistics of houses inland with houses at ocean proximity
## in order to understand behaviour of the people
######################################################################################
```

```python
#################################################
## We investigate houses ocean proximity
#################################################
```

```python
house_ocean = houses.loc[houses['ocean_proximity']=='NEAR BAY']
#print(house_ocean)
print('Number of houses at ocean_proximity is:', len(house_ocean))
```

```python
print('mean for median_income for houses near ocean is:', house_ocean [['median_income']].mean())
print('')
print('mean for median_house_value for houses near ocean is:',house_ocean[['median_house_value']].mean())
print('')
print('mean for housing_mean_age for houses near ocean is:', house_ocean[['housing_median_age']].mean())
print('')
print('mean for total_rooms for houses near ocean is:', house_ocean[['total_rooms']].mean())
```

```python
house_ocean_sorted = house_ocean.sort_values(by='median_house_value', ascending=False)
#print(house_ocean_sorted)
```

```python
richest_house_ocean_sorted = house_ocean_sorted.loc[house_ocean_sorted['median_house_value'] >= 400000]
#print('how many houses at ocen proximity are more expensive then 40000 USD are:',  richest_house_ocean_sorted)
print('Total number of houses at ocean side that are more expensive then 400000 USD are:  is:',len(richest_house_ocean_sorted))
```

```python
print('mean for median_income for houses near ocean with median_house_value greater than 400000 USD is:', richest_house_ocean_sorted[['median_income']].mean())
print('')
print('mean for median_house_value for houses near ocean with median_house_value greater than 400000 USD  is:',richest_house_ocean_sorted[['median_house_value']].mean())
print('')
print('mean for housing_mean_age for houses near ocean with median_house_value greater than 400000 USD  is:', richest_house_ocean_sorted[['housing_median_age']].mean())
print('')
print('mean for total_rooms for houses near ocean with median_house_value greater than 400000 USD  is:', richest_house_ocean_sorted[['total_rooms']].mean())
```

```python
#################################################
## We investigate houses Inland proximity
#################################################
```

```python
house_inland = houses.loc[houses['ocean_proximity']=='INLAND']
#print(house_inland)
print('Number of houses at inland is:', len(house_inland))
```

```python
print('mean for median_income for houses inland is:', house_inland[['median_income']].mean())
print('')
print('mean for median_house_value for houses inland is:',house_inland[['median_house_value']].mean())
print('')
print('mean for housing_mean_age for houses inland is:', house_inland[['housing_median_age']].mean())
print('')
print('mean for total_rooms for houses inland is:', house_inland[['total_rooms']].mean())
```

```python
house_inland_sorted = house_inland.sort_values(by='median_house_value', ascending=False)
#print(house_inland_sorted)
```

```python
richest_house_inland_sorted = house_inland_sorted.loc[house_inland_sorted['median_house_value'] >= 400000]
#print('how many houses at inland are more expensive then 40000 USD are:',  richest_house_inland_sorted)
print('Total number of houses at inland that are more expensive then 40000 USD are:  is:',len(richest_house_inland_sorted))
```

```python
print('mean for median_income for houses inland with median_house_value greater than 400000 USD is:', richest_house_inland_sorted[['median_income']].mean())
print('')
print('mean for median_house_value for houses inland with median_house_value greater than 400000 USD  is:',richest_house_inland_sorted[['median_house_value']].mean())
print('')
print('mean for housing_mean_age for houses inland with median_house_value greater than 400000 USD  is:', richest_house_inland_sorted[['housing_median_age']].mean())
print('')
print('mean for total_rooms for houses inland with median_house_value greater than 400000 USD  is:', richest_house_inland_sorted[['total_rooms']].mean())
```
